```python
import zipfile
import os

# Specify the ZIP file name and target directory
zip_file = "upload folder.zip"  # Replace with your ZIP file name
extract_dir = "datasets"  # Replace with your target directory

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Files extracted to: {extract_dir}")

# List extracted files (optional)
print("Extracted files:", os.listdir(extract_dir))
```

```python
import os

label_dir = "/mnt/local/o.bobylev/datasets/upload folder/train"  # Path to labels
max_class = 2  # Max allowed class index (since nc=3 â†’ indices 0,1,2)

for label_file in os.listdir(label_dir):
    with open(os.path.join(label_dir, label_file), 'r') as f:
        lines = f.readlines()
        for line in lines:
            class_idx = int(line.split()[0])
            if class_idx > max_class:
                print(f"Invalid class {class_idx} in file: {label_file}")
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model with 2 GPUs
results = model.train(data="dataset.yaml", epochs=10, imgsz=1280, device=[0, 1], batch=32)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Run inference
results = model.predict(
    source="datasets/upload folder/test/document_1535.jpeg",
    conf=0.7,
    batch=16,
    imgsz=1280,
    save=False,  # Disable saving
    device=1
)
annotated_img = results[0].plot(
    labels=True, 
    conf=True,    
    line_width=5,  # Thinner bounding boxes (optional)
    font_size=10  # Smaller font (if any text remains)
)

# Display
plt.figure(figsize=(20, 20))
plt.imshow(annotated_img)
plt.axis('off')
plt.show()
```
